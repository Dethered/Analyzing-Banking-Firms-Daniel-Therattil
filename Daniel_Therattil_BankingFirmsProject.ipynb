## Import libraries needed to perform webscraping and graphing:

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from io import StringIO 
```

# Company #1: JPMorgan Chase & Co's

## To webscrape JPMorgan Chase & Co's annual revenue perform the following:

```python
url = "https://www.macrotrends.net/stocks/charts/JPM/jpmorgan-chase/revenue" #Create a variable assigned to the string of the JPMorgan annual revenue data.

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/91.0.4472.124 Safari/537.36"
} #Used to stop 403 errors from webscraping by mimicking regular webtraffic.

response = requests.get(url, headers=headers).text #Sends an HTTP "GET" request to retrieve data from the url variable I defined.
response_wrapped = StringIO(response) #Converts response from a raw HTML string to an I/O object which will help the pandas library simplifying the process for pandas to parse the HTML content  effectively
```

## To webscrape tables the following was performed the using the pandas and BeautifulSoup libraries:

```python
JPM_dflist = pd.read_html(response_wrapped, flavor='bs4') #List of JP Morgan dataframes obtained using pandas library.
JPM_df = JPM_dflist[0] #Selected the JPM data frame displaying its Annual Revenue.
JPM_df.columns = ['Year','Annual Revenue'] #Renamed the columns for easier managment.
```

## Clean up messy data by omitting conflicting strings, converting conflicting data types, and transforming dataframe to be numeric in preparation for graphing:

```python
JPM_df['Annual Revenue'] = JPM_df['Annual Revenue'].str.replace(',','') #To get rid of the commas stopping me from converting it to a float, which is needed for me to graph it.
JPM_df['Annual Revenue'] = JPM_df['Annual Revenue'].str.replace('$','') #To get rid of the dollar sign stopping me from converting it to a float, which is needed for me to graph it.
JPM_df['Annual Revenue'] = pd.to_numeric(JPM_df['Annual Revenue']) #Annual Revenue was originally an "object" type hence it would not be able to be graphed unless I changed it to a numeric data type.
```

## Finally graph the data using matplotlib:

```python
JPM_graph = JPM_df.plot(x = 'Year', y = 'Annual Revenue', color='black', linestyle='dotted')
plt.show()
```

# **Revenue Growth Rate:**

## Growth Rate of JPMorgan Chase & Co:

## Import libraries needed to find and visualize the growth rate of annual revenue:

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
from io import StringIO 
```

## Run following before finding growth rate:

```python
url = "https://www.macrotrends.net/stocks/charts/JPM/jpmorgan-chase/revenue"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/91.0.4472.124 Safari/537.36"
} 

response = requests.get(url, headers=headers).text 
response_wrapped = StringIO(response)

JPM_dflist = pd.read_html(response_wrapped, flavor='bs4')
JPM_df = JPM_dflist[0] 
JPM_df.columns = ['Year','Annual Revenue']

JPM_df['Annual Revenue'] = JPM_df['Annual Revenue'].str.replace(',','') 
JPM_df['Annual Revenue'] = JPM_df['Annual Revenue'].str.replace('$','') 
JPM_df['Annual Revenue'] = pd.to_numeric(JPM_df['Annual Revenue']) 
```

## Find growth rate of annual revenue:

```python
JPM_growth_rate = JPM_df['Annual Revenue'] = JPM_df['Annual Revenue'].pct_change().mul(100).round(1) #Calculates the percentage change(pct) and growth rate of revenue rounded to the tenths place. 
print(JPM_growth_rate)

JPM_growth_rate_sum = JPM_growth_rate.sum()
JPM_growth_rate_avg = JPM_growth_rate_sum / len(JPM_growth_rate)

print(JPM_growth_rate_avg)
```

## Growth Rate Visualized:

```python
JPM_growth_rate_graph = JPM_growth_rate.plot(x = 'Year', y = 'Annual Revenue', color='red', linestyle='--')
JPM_growth_rate_graph.set_xlabel("Year")
JPM_growth_rate_graph.set_ylabel("Percent change in annual revenue")
plt.show()
```

Originally, it seemed that JPMorgan Chase & Co. was growing at an exponentially high rate with no stop but after analyzing it's growth rate we can see it had an expolsive growth before plateauing out to a slow increase in annual revenue growth.
